```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "pseudolabels.csv"

df = pd.read_csv(filepath)
df.head()
```

```python
df.sample(n=2, axis="index")
```

```python
for _, row in df.sample(10, axis="index").iterrows():
    print(row[1:].sum())
```

```python
plt.figure(figsize=(20, 20))
n=10
for i, (sample, row) in enumerate(df.sample(n, axis="index").iterrows()):
    plt.subplot(n//2, 2, i+1)
    plt.bar(range(206), height=row[1:]/row[1:].sum())
    plt.ylim(0,0.2)
plt.tight_layout()
plt.show()
```
